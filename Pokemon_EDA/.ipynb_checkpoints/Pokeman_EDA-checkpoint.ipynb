# Import data manipulation and visualization libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

# Loading the Pokemon dataset

```python
df_raw = pd.read_csv("pokemon.csv")
df_raw
```

# Create a copy of the raw dataset for manipulation

```python
df = df_raw.copy()
df.tail()
```

```python
# Checking the dimensions of the dataset (rows, columns)
df.shape
```

```python
# Geting information about the dataset (data types, non-null values)
df.info()
```

```python
# Geting the column names of the dataset
df.columns
```

```python
# Geting descriptive statistics of numerical features
df.describe().round(2)
```

```python
# Geting descriptive statistics of categorical and boolean features
df.describe(include=("O","bool"))
```

```python
# Checking for duplicate Pokemon names
df[df["Name"].duplicated()]
```

```python
# Checking for completely duplicate rows
df[df.duplicated()]
```

```python
# Filling missing values in 'Type 2' column with 'None'
df["Type 2"].fillna(value="None",inplace=True)
df.sample() # Display a random sample of rows
```

```python
# Counting the number of non-null values in each column
df.count()
```

```python
# Converting column names to lowercase
df.columns=df.columns.str.lower()
df.columns
```

```python
# Clean and standardize column names
cols=[]
for i in df.columns:
    if " " in i:
        i=i.replace(" ","_")
    if "." in i:
        i=i.replace(".","")
    elif "#" == i:
        i = "Num"
    cols.append(i)
df.columns=cols
df.columns
```

## Visualization

# Create a countplot of Pokemon types

```python
sns.catplot(y="type_1",data=df,kind="count",order=df["type_1"].value_counts().index,aspect=2).fig.suptitle("count of type1 pokemon",y=1.0,color="red")
```

```python
sns.catplot(x="type_2",data=df,kind="count",order=df["type_2"].value_counts().index,aspect=2).fig.suptitle("Count of tyoe2 pokemon",y=1,color="purple")
```

# Create a countplot of Pokemon generations with annotations

```python

ax=sns.countplot(x="generation",data=df)
for i in ax.patches:
    x=i.get_x()
    y=i.get_height()/2
    ax.annotate(y,(x+0.2,y))
```

## Visualization for generation and Legendry Counts

```python
sns.catplot(x="generation",data=df,kind="count",hue="legendary").fig.suptitle("Count of Lengendry Pokemon",y=1.03)
plt.show()
```

```python
ax=sns.countplot(x="legendary",data=df,color="yellow",width=0.5)
for i in ax.patches:
    x=i.get_x()
    y=i.get_height()
    ax.annotate(y,(x+0.2,y/2),color="red",ha="center")# ha=horigentally aligned
ax.set_title("Count of Legendry",color="blue")
```
